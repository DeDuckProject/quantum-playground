calculating the statevector at the end of a circuit

```python
# playing with statevector:
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi

qreg_q = QuantumRegister(2, 'q')

circuit = QuantumCircuit(qreg_q)

circuit.h(0)
circuit.cx(qreg_q[0],qreg_q[1])
circuit.h(0)

circuit.draw()

state = Statevector.from_int(0, 2**2)

# Evolve the state by the quantum circuit
state = state.evolve(circuit)

#draw using latex
state.draw('latex')
```

Playing with bloch spheres

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline

qreg_q = QuantumRegister(2, 'q')

circuit = QuantumCircuit(qreg_q)

circuit.h(0)
# circuit.cx(qreg_q[0],qreg_q[1])
circuit.h(1)

state = Statevector.from_instruction(circuit)
plot_bloch_multivector(state, title="New Bloch Multivector", reverse_bits=False)
```
