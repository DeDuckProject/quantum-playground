# Solovey-Kitaev
This is a demo of Solovey-Kitaev building blocks, not an implementation of their algorithm

```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from qiskit.quantum_info import Statevector
from qiskit.quantum_info import DensityMatrix
```

```python
# Playground for showing gates as matrix operators
from math import pi
circuit = QuantumCircuit(1)

# first vector:
# tempCircuit.h(0)
# tempCircuit.t(0)
# tempCircuit.h(0)
# tempCircuit.t(0)

# seconds vector
circuit.t(0)
circuit.h(0)
circuit.t(0)
circuit.h(0)

circuit.save_unitary()
usim = Aer.get_backend('aer_simulator')
unitary = usim.run(circuit).result().get_unitary()
array_to_latex(unitary, prefix="\\text{U = }\n")
```

```python
circuit = QuantumCircuit(1)

arbitrary_code = "1001011011101"

for s in arbitrary_code:
    if s=="0":
        # first vector:
        circuit.h(0)
        circuit.t(0)
        circuit.h(0)
        circuit.t(0)
    else:
        # seconds vector
        circuit.t(0)
        circuit.h(0)
        circuit.t(0)
        circuit.h(0)

state = Statevector.from_instruction(circuit)
plot_bloch_multivector(state, title="New Bloch Multivector", reverse_bits=False)
```
